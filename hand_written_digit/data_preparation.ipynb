```python
import os
import shutil

save_folder = "/gdrive/MyDrive/hand_written_digit_4/all_images"  # Thay bằng thư mục bạn muốn lưu ảnh
os.makedirs(save_folder, exist_ok=True)

for i, img_path in enumerate(image_lists):
    # Lấy tên file gốc
    file_name = os.path.basename(img_path)
    # Tạo tên mới tránh trùng (ví dụ: 000001_img.png)
    new_name = f"{i:06d}_{file_name}"
    dst_path = os.path.join(save_folder, new_name)
    # Copy ảnh sang folder mới
    shutil.copy2(img_path, dst_path)

print(f"Đã copy {len(image_lists)} ảnh vào {save_folder}")
```

```python
import os
import shutil

all_images_folder = "/gdrive/MyDrive/hand_written_digit_4/all_images"  # thư mục chứa tất cả ảnh
dst_root = "/gdrive/MyDrive/hand_written_digit_4/sort_images"       # thư mục gốc để phân chia theo nhãn
os.makedirs(dst_root, exist_ok=True)

# Tạo thư mục 0..9 nếu chưa có
for i in range(10):
    os.makedirs(os.path.join(dst_root, str(i)), exist_ok=True)

# Duyệt ảnh trong all_images_folder
for img_file in os.listdir(all_images_folder):
    if img_file.lower().endswith(('.png', '.jpg', '.jpeg')):
        # Lấy ký tự thứ 8 (index 7)
        label_char = img_file[7]

        # Kiểm tra ký tự có phải số từ 0-9 không
        if label_char.isdigit():
            label_folder = os.path.join(dst_root, label_char)
            src_path = os.path.join(all_images_folder, img_file)
            dst_path = os.path.join(label_folder, img_file)
            shutil.copy2(src_path, dst_path)

print("✅ Đã phân loại xong ảnh theo nhãn vào các thư mục 0..9.")
```

```python
import os
import cv2
import matplotlib.pyplot as plt

dst_root = "/gdrive/MyDrive/hand_written_digit_4/sort_images"

plt.figure(figsize=(15, 4))
for i in range(10):
    label_folder = os.path.join(dst_root, str(i))
    img_files = [f for f in os.listdir(label_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    if img_files:
        img_path = os.path.join(label_folder, img_files[0])
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        plt.subplot(2, 5, i + 1)
        plt.imshow(img, cmap='gray')
        plt.title(f'Label: {i}')
        plt.axis('off')

plt.tight_layout()
plt.show()
```
