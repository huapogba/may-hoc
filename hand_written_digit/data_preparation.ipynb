```python
import os
import shutil

all_images_folder = "/gdrive/MyDrive/hand_written_digit_4/all_images"  # th∆∞ m·ª•c ch·ª©a t·∫•t c·∫£ ·∫£nh
dst_root = "/gdrive/MyDrive/hand_written_digit_4/sort_images"       # th∆∞ m·ª•c g·ªëc ƒë·ªÉ ph√¢n chia theo nh√£n
os.makedirs(dst_root, exist_ok=True)

# T·∫°o th∆∞ m·ª•c 0..9 n·∫øu ch∆∞a c√≥
for i in range(10):
    os.makedirs(os.path.join(dst_root, str(i)), exist_ok=True)

# Duy·ªát ·∫£nh trong all_images_folder
for img_file in os.listdir(all_images_folder):
    if img_file.lower().endswith(('.png', '.jpg', '.jpeg')):
        # L·∫•y k√Ω t·ª± th·ª© 8 (index 7)
        label_char = img_file[7]

        # Ki·ªÉm tra k√Ω t·ª± c√≥ ph·∫£i s·ªë t·ª´ 0-9 kh√¥ng
        if label_char.isdigit():
            label_folder = os.path.join(dst_root, label_char)
            src_path = os.path.join(all_images_folder, img_file)
            dst_path = os.path.join(label_folder, img_file)
            shutil.copy2(src_path, dst_path)

print("‚úÖ ƒê√£ ph√¢n lo·∫°i xong ·∫£nh theo nh√£n v√†o c√°c th∆∞ m·ª•c 0..9.")
```

```python
import os
import cv2
import matplotlib.pyplot as plt

dst_root = "/gdrive/MyDrive/hand_written_digit_4/sort_images"

plt.figure(figsize=(15, 4))
for i in range(10):
    label_folder = os.path.join(dst_root, str(i))
    img_files = [f for f in os.listdir(label_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    if img_files:
        img_path = os.path.join(label_folder, img_files[0])
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        plt.subplot(2, 5, i + 1)
        plt.imshow(img, cmap='gray')
        plt.title(f'Label: {i}')
        plt.axis('off')

plt.tight_layout()
plt.show()
```
